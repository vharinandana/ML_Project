```python
df['balance'] = df['balance'].fillna(df['balance'].median())
```

```python
df['salary_balance_ratio'] = df['salary_balance_ratio'].fillna(df['salary_balance_ratio'].median())
```

```python
df['estimated_salary'] = df['estimated_salary'].fillna(df['estimated_salary'].mean())
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import OneHotEncoder

onh = OneHotEncoder(sparse_output=False)
data = onh.fit_transform(df[['country']])
df1 = pd.DataFrame(data,columns=onh.get_feature_names_out())
```

```python
onh1 = OneHotEncoder(sparse_output=False)
data1 = onh1.fit_transform(df[['gender']])
df2 = pd.DataFrame(data1,columns = onh1.get_feature_names_out())
```

```python
df = pd.concat([df,df1,df2],axis=1)
```

```python
df = df.drop(['country','gender'],axis=1)
```

```python
df['tenure'] = df['tenure'].replace({'zero':0,'one':1,'two':2,'three':3,'four':4,'five':5,'six':6,'seven':7,'eight':8,'nine':9,'ten':10}).astype(int)
```

```python
df.corr()['churn']
```

```python
df.drop(['customer_id'],axis=1,inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
x = df.drop('churn',axis = 1)
y = df['churn']
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0,stratify=y)
```

```python
from scipy.stats import randint

parameters= {'n_estimators': randint(10, 100),'criterion':["gini", "entropy", "log_loss"],'max_depth': randint(3, 10),'min_samples_split': randint(2, 10),'max_features': randint(1, 6)}
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV

rfc = RandomForestClassifier()
rscv = RandomizedSearchCV(rfc,param_distributions=parameters,cv=10,random_state=42)
rscv.fit(x_train,y_train)
rscv.best_params_
```

```python
rfc = RandomForestClassifier(criterion = 'log_loss',max_depth =8, max_features = 5, min_samples_split = 3, n_estimators = 97,class_weight='balanced')
rfc.fit(x_train,y_train)
y_pred = rfc.predict(x_test)
y_pred
```

```python
training_score = rfc.score(x_train,y_train)
print('training score =',training_score)
testing_score = rfc.score(x_test,y_test)
print('testing score =',testing_score)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,y_pred))
```

```python
print("Confusion Matrix")
print(confusion_matrix(y_test,y_pred))
print("\nAccuracy =",accuracy_score(y_test,y_pred))
print()
print(classification_report(y_test,y_pred))
```

```python
y_train_pred = rfc.predict(x_train)
y_train_pred
```

```python
print("Confusion Matrix")
print(confusion_matrix(y_train,y_train_pred))
print("\nAccuracy =",accuracy_score(y_train,y_train_pred))
print()
print(classification_report(y_train,y_train_pred))
```
